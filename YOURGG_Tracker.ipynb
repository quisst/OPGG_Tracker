### BeautifulSoup으로 n인분, 팀운, 라인전, 킬, 데스, 어시스트 크롤링

```python
driver = webdriver.Chrome('C:\\JupyterLab\\chromedriver.exe')
driver.get('https://your.gg/kr/profile/{}?mc=SoloRank'.format(nickname))

WebDriverWait(driver, 2).until(EC.element_to_be_clickable((By.ID, "profileUpdateBtn")))
driver.find_element_by_id('profileUpdateBtn').click()

# --------------------------------------------------------------------------------------

html_code = driver.page_source
driver.close()
html = BeautifulSoup(html_code, 'html.parser')

matchlist_list = html.select('.gg-ggx-on-matchlist a')
K_As = html.select('.text-success')
_D_s = html.select('.text-warning')
game_result_div = html.select('body>div>div.container-fluid.page-body-wrapper>div>div>div.row.mt-3>div.col-lg-8.col-12>div.d-flex.flex-column.mt-3>div:nth-child(1)>div> div>div.d-flex.flex-column.justify-content-between.align-items-center>div')

in_boon_list, team_luck_list, line_fight_list, kills, deaths, assists, champions, game_results = [], [], [], [], [], [], [], []

for i in range(0, 3 * spin, 3):
    in_boon = matchlist_list[i].get_text().strip()
    team_luck = matchlist_list[i+1].get_text().strip()
    line_fight = matchlist_list[i+2].get_text().strip()
    
    in_boon_list.append(in_boon)
    team_luck_list.append(team_luck)
    line_fight_list.append(line_fight)
    
for i in range(2, 2 * spin + 2, 2):
    kill = int(K_As[i].get_text())
    assist = int(K_As[i+1].get_text())
    
    kills.append(kill)
    assists.append(assist)
    
for i in range(3, spin + 3):
    death = int(_D_s[i].get_text())
    deaths.append(death)
    
for i in range(spin):
    champion_pre = html.select('.gg-matchlist-champion')[i]
    champion = champion_pre["alt"].split(' ')[1]
    champions.append(champion)

for i in range(1, 1 + spin):
    game_result_div = html.select('body>div>div.container-fluid.page-body-wrapper>div>div>div.row.mt-3>div.col-lg-8.col-12>div.d-flex.flex-column.mt-3>div:nth-child({})>div> div>div.d-flex.flex-column.justify-content-between.align-items-center>div'.format(i))
    str(game_result_div).split("</div>")
    game_results.append(game_result_div[2].get_text().strip())
    
in_boon_list, team_luck_list, line_fight_list, kills, deaths, assists, champions, game_results
```

### 데이터프레임화

```python
index = []
for i in range(spin):
    index.append('Game {0}'.format(i+1))
    
pre_df = [[0 for col in range(8)] for row in range(spin)]

for i in range(spin):
    pre_df[i][0] = kills[i]
    pre_df[i][1] = deaths[i]
    pre_df[i][2] = assists[i]
    pre_df[i][3] = in_boon_list[i]
    pre_df[i][4] = team_luck_list[i]
    pre_df[i][5] = line_fight_list[i]
    pre_df[i][6] = champions[i]
    pre_df[i][7] = game_results[i]

df = pd.DataFrame(pre_df, columns=['Kills', 'Deaths', 'Assists', '인분', '팀운', '라인전', 'champion', 'game_result'], index=index)
df
```
